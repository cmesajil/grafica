```julia
# Crea un modelo
model = Model(Gurobi.Optimizer)

# Define las variables
@variable(model, x >= 0)
@variable(model, y >= 0)

# Define la función objetivo
@objective(model, Max, 2x + 3y)

# Define las restricciones
@constraint(model, x + y <= 5)
@constraint(model, 2x + y <= 8)

# Resuelve el modelo
optimize!(model)

# Imprime la solución
println("Valor óptimo: ", objective_value(model))
println("x = ", value(x))
println("y = ", value(y))
```
